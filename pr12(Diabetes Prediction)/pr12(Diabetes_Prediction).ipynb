````python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
````

##Data collection and Analysis
##PIMA diabetes dataset

````python
x=diabetes_dataset.drop(columns="Outcome",axis=1)
y=diabetes_dataset["Outcome"]
````

````python
scaler=StandardScaler()
x=scaler.fit_transform(x)
````

````python
x
````

````python
y
````

````python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42,stratify=y)
````

````python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
````

#Model Training-->SVC#

````python
Support_vector_machine=SVC(kernel="rbf")
Support_vector_machine.fit(x_train,y_train)
````

````python
train_pred=Support_vector_machine.predict(x_train)
print("accuracy_train :",accuracy_score(train_pred,y_train))
````

````python
test_pred=Support_vector_machine.predict(x_test)
print("accuracy_test :",accuracy_score(test_pred,y_test))
````

##Make a Predicatve System##

````python
input_data = (1,	89,	66,	23,	94,	28.1,	0.167,	21)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
# standardize the input data
std_data = scaler.transform(input_data_reshaped)
prediction = Support_vector_machine.predict(std_data)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
````
